```python
import pandas as pd
```

```python
state_info_2017_df = pd.read_csv("ACS_17_5YR_S0102_with_ann.csv", skiprows=[0])
state_info_2016_df = pd.read_csv("ACS_16_5YR_S0102_with_ann.csv", skiprows=[0])
state_info_2015_df = pd.read_csv("ACS_15_5YR_S0102_with_ann.csv", skiprows=[0])
state_info_2014_df = pd.read_csv("ACS_14_5YR_S0102_with_ann.csv", skiprows=[0])
state_info_2013_df = pd.read_csv("ACS_13_5YR_S0102_with_ann.csv", skiprows=[0])

state_info_2013_df.head()
```

```python
gender_state_cols_2017_df = state_info_2017_df.filter(regex="Estimate")
gender_state_2017_df = state_info_2017_df.filter(regex="Geography")
gender_state_cols_2016_df = state_info_2016_df.filter(regex="Estimate")
gender_state_2016_df = state_info_2016_df.filter(regex="Geography")
gender_state_cols_2015_df = state_info_2015_df.filter(regex="Estimate")
gender_state_2015_df = state_info_2015_df.filter(regex="Geography")
gender_state_cols_2014_df = state_info_2014_df.filter(regex="Estimate")
gender_state_2014_df = state_info_2014_df.filter(regex="Geography")
gender_state_cols_2013_df = state_info_2013_df.filter(regex="Estimate")
gender_state_2013_df = state_info_2013_df.filter(regex="Geography")

gender_state_cols_2013_df.head()
gender_state_2013_df.head()
```

```python
total_gender_state_2017_df = pd.concat([gender_state_2017_df, gender_state_cols_2017_df], axis=1)
total_gender_state_2016_df = pd.concat([gender_state_2016_df, gender_state_cols_2016_df], axis=1)
total_gender_state_2015_df = pd.concat([gender_state_2015_df, gender_state_cols_2015_df], axis=1)
total_gender_state_2014_df = pd.concat([gender_state_2014_df, gender_state_cols_2014_df], axis=1)
total_gender_state_2013_df = pd.concat([gender_state_2013_df, gender_state_cols_2013_df], axis=1)

total_gender_state_2013_df.head()
```

```python
gender_state_2017_data = total_gender_state_2017_df[["Geography", "Total; Estimate; Total population", "Total; Estimate; SEX AND AGE - Male", "Total; Estimate; SEX AND AGE - Female"]]
gender_state_2016_data = total_gender_state_2016_df[["Geography", "Total; Estimate; Total population", "Total; Estimate; SEX AND AGE - Male", "Total; Estimate; SEX AND AGE - Female"]]
gender_state_2015_data = total_gender_state_2015_df[["Geography", "Total; Estimate; Total population", "Total; Estimate; SEX AND AGE - Male", "Total; Estimate; SEX AND AGE - Female"]]
gender_state_2014_data = total_gender_state_2014_df[["Geography", "Total; Estimate; Total population", "Total; Estimate; SEX AND AGE - Male", "Total; Estimate; SEX AND AGE - Female"]]
gender_state_2013_data = total_gender_state_2013_df[["Geography", "Total; Estimate; Total population", "Total; Estimate; SEX AND AGE - Male", "Total; Estimate; SEX AND AGE - Female"]]

gender_state_2013_data.head()
```

```python
gender_state_2017_data.columns
gender_state_2016_data.columns
gender_state_2015_data.columns
gender_state_2014_data.columns
gender_state_2013_data.columns
```

```python
gender_state_2017_df = gender_state_2017_data.rename(columns={"Geography": "State", "Total; Estimate; Total population": "Total Population 2017",
       "Total; Estimate; SEX AND AGE - Male": "Total Male Population 2017",
       "Total; Estimate; SEX AND AGE - Female": "Total Female Population 2017"})

gender_state_2016_df = gender_state_2016_data.rename(columns={"Geography": "State", "Total; Estimate; Total population": "Total Population 2016",
       "Total; Estimate; SEX AND AGE - Male": "Total Male Population 2016",
       "Total; Estimate; SEX AND AGE - Female": "Total Female Population 2016"})

gender_state_2015_df = gender_state_2015_data.rename(columns={"Geography": "State", "Total; Estimate; Total population": "Total Population 2015",
       "Total; Estimate; SEX AND AGE - Male": "Total Male Population 2015",
       "Total; Estimate; SEX AND AGE - Female": "Total Female Population 2015"})

gender_state_2014_df = gender_state_2014_data.rename(columns={"Geography": "State", "Total; Estimate; Total population": "Total Population 2014",
       "Total; Estimate; SEX AND AGE - Male": "Total Male Population 2014",
       "Total; Estimate; SEX AND AGE - Female": "Total Female Population 2014"})

gender_state_2013_df = gender_state_2013_data.rename(columns={"Geography": "State", "Total; Estimate; Total population": "Total Population 2013",
       "Total; Estimate; SEX AND AGE - Male": "Total Male Population 2013",
       "Total; Estimate; SEX AND AGE - Female": "Total Female Population 2013"})

gender_state_2013_df.head()
```

```python
total_gender_2013_2014_data = pd.merge(gender_state_2013_df, gender_state_2014_df, on=["State"])
total_gender_2013_2014_2015_data = pd.merge(total_gender_2013_2014_data, gender_state_2015_df, on=["State"])
total_gender_2013_2014_2015_2016_data = pd.merge(total_gender_2013_2014_2015_data, gender_state_2016_df, on=["State"])
total_gender_state_data = pd.merge(total_gender_2013_2014_2015_2016_data, gender_state_2017_df, on=["State"])

total_gender_state_data.head()
```
