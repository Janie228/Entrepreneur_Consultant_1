```python
%matplotlib notebook

# Dependencies
import os
import sys
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import pandas as pd
import shutil
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
# Set directory file paths
output_path = os.path.join(".", "Output", "data_clean_output.txt")
filepath = os.path.join(".", "Output", "clean_Companies_total.csv")
```

```python
business_data = pd.read_csv(filepath)
state_list = business_data["State"].values.tolist()
business_df = business_data.set_index("State")
# Create a list of the years that we will use as our x axis
years = [2012,2013,2014,2015,2016]

for item in state_list:
#     print(item)
    num_firm_yr = business_df.loc[item,["2012_Num_Companies","2013_Num_Companies", "2014_Num_Companies", "2015_Num_Companies", "2016_Num_Companies"]]
    plt.plot(years, (num_firm_yr/1000), marker="o", label=item)

fig = plt.gcf()
fig.set_size_inches(10, 15)
plt.grid()
plt.title("Total Companies by State and Year")
plt.xlabel("Years")
plt.ylabel("Total Companies (Thousand)")
plt.xlim(round(min(years) - 1), round(max(years) + 1.5))
plt.ylim(min(business_df["2013_Num_Companies"]/1000) - 25, max(business_df["2016_Num_Companies"]/1000 + 75))
# Create a legend
plt.legend(title="States", loc='right')
        
# Save the plot and display it
plt.savefig("Images/Total_Company.png")
plt.show()
```
