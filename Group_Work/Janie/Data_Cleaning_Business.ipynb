```python
# Dependencies
import os
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'

# Load excel
filepath = os.path.join(".", "Resources/State/Business", "state_naicssector_2015.xlsx")
print(filepath)

business_data = pd.read_excel(filepath, skiprows=5)
business_data.head()
```

```python
business_data = business_data.drop(business_data.index[[0,1]])
splitted_col = business_data["ENTERPRISE EMPLOYMENT SIZE"].str.split(":", n = 1, expand = True) 
# Split enterprise employment size to # & size cols
business_data["Num"]= splitted_col[0].str.strip()   
business_data["Company_Size"]= splitted_col[1].str.strip()
business_data.head(10)
```

```python
state_data = business_data[["STATE DESCRIPTION", "NUMBER OF FIRMS", "NAICS DESCRIPTION", "Num", "Company_Size"]]
state_data.columns = ["State", "Num_Companies", "NAICS_Desc", "Num(#)", "Company_Size"]
state_data["Num_Companies"] = state_data["Num_Companies"].astype(int)
state_data.head(15)
```

```python
state_total = state_data.loc[(state_data["NAICS_Desc"] == "Total") & (state_data["Company_Size"] == "Total")]
state_total.count()
```

```python
state_detail = state_data.loc[(state_data["NAICS_Desc"] == "Total") & (~state_data["Company_Size"].str.contains('<')) & (state_data["Company_Size"] != "Total")]
company_total = state_detail.groupby("State")
company_total = company_total["Num_Companies"].sum()

state_detail["Company_Size%"] =  [round((state["Num_Companies"]/company_total[state["State"]]) * 100,1)  for i, state in state_detail.iterrows()]
state_Co_df = state_detail[["State", "Num_Companies", "Company_Size%", "Company_Size"]]

state_Co_df.head()
```

```python
co_ttl_industry = state_data.loc[(state_data["NAICS_Desc"] != "Total") & (state_data["Company_Size"] == "Total")]
co_ttl_industry = co_ttl_industry.reset_index(drop=True)
co_ttl_industry = co_ttl_industry[["State", "Num_Companies", "NAICS_Desc"]].rename(columns={"NAICS_Desc": "Industry"})
co_ttl_industry.head(20)
```

```python
co_ind_grp = state_data.loc[(state_data["NAICS_Desc"] != "Total") & (~state_data["Company_Size"].str.contains('<')) & (state_data["Company_Size"] != "Total")]
co_ind_grp = co_ind_grp.reset_index(drop=True)
co_ind_grp  = co_ind_grp[["State", "Num_Companies", "Company_Size", "NAICS_Desc"]].rename(columns={"NAICS_Desc": "Industry"})
co_ind_grp .head(20)
```
