```python
# Dependencies
import json
import os
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'

# Load JSON
filepath = os.path.join(".", "Resources/Business_States_2012-2016")
x = "state_naicssector_2016.xlsx"
# print(filepath)
yr = "2016"
print(f'{filepath}/{x}, {yr}')
```

```python
business_data = pd.read_excel(f'{filepath}/{x}' , skiprows=5)
# business_data.head()
business_data = business_data.drop(business_data.index[[0,1]])
splitted_col = business_data["ENTERPRISE EMPLOYMENT SIZE"].str.split(":", n = 1, expand = True) 

# Split enterprise employment size to # & size cols
business_data["Num"]= splitted_col[0].str.strip()   
business_data["Firm_Size"]= splitted_col[1].str.strip()
state_data = business_data[["STATE DESCRIPTION", "NUMBER OF FIRMS", "NAICS DESCRIPTION", "Num", "Firm_Size"]]
state_data.columns = ["State", "Num_Firms", "NAICS_Desc", "Num(#)", "Firm_Size"]
state_data["Num_Firms"] = state_data["Num_Firms"].astype(int)

state_data = business_data[["STATE DESCRIPTION", "NUMBER OF FIRMS", "NAICS DESCRIPTION", "Num", "Firm_Size"]]
state_data.columns = ["State", yr + "_Num_Firms", "NAICS_Desc", "Num(#)", "Firm_Size"]
state_data[yr + "_Num_Firms"] = state_data[yr + "_Num_Firms"].astype(int)

state_total = state_data.loc[(state_data["NAICS_Desc"] == "Total") & (state_data["Firm_Size"] == "Total")]
yr = state_total[["State", yr + "_Num_Firms"]]
                    
yr
```
