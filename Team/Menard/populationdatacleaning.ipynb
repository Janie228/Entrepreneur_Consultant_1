```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv('population_clean_data.csv')
state_list = df["States"].values.tolist()
df = df.set_index("States")
# Create a list of the years that we will use as our x axis
years = [2013,2014,2015,2016,2017]
for item in state_list:
#     print(item)
   num_population = df.loc[item,["2013","2014", "2015", "2016", "2017"]]
   plt.plot(years, (num_population/1000), marker="o", label=item)
fig = plt.gcf()
fig.set_size_inches(10, 15)
plt.grid()
plt.title("Total population by State and Year")
plt.xlabel("Years")
plt.ylabel("Total population (Thousand)")
plt.xlim(round(min(years) - 1), round(max(years) + 1.5))
plt.ylim(min(df["2013"]/1000) - 25, max(df["2017"]/1000 + 75))
# Create a legend
plt.legend(title="States", loc='right')
# Save the plot and display it
plt.savefig("Images/Total_Firm.png")
plt.show()
```
