```python
# Dependencies
import os
import pandas as pd
import numpy as py
pd.options.mode.chained_assignment = None  # default='warn'

# Load excel
filepath = os.path.join(".", "Resources/State/Crimes", "Table_69_Arrest_by_State_2011.xls")
output_crime_ttl = os.path.join(".", "Output/Crime", "crime_ttl_2011.csv")
output_crime18_ttl = os.path.join(".", "Output/Crime", "crime_18ttl_2015.csv")
```

```python
# #Copy odd # row state col to even # row null state col
for index, row in crime_data.iterrows():
    if index == 0:
        state = row["State"]
    else:
        if index % 2 == 0:
            state = row["State"]
        else:
            crime_data["State"][index] = state            

crime_data["State"] = crime_data.State.str.capitalize()   # Format state to readable

namesList = ["Type", "Total_Crimes", "Violent", "Property", "Murder", "Rape", "Robery", "Assault", "Burglary", "Theft", "Vehicle_Theft", "Arson", "Other_Assault", "Forgery", "Fraud", "Embezzlement", "Stolen_Property", "Vandalism", "Weapons", "Prostitution", "Sex_Offense", "Drug_Abuse", "Gambling", "Family_Offenses", "Driving_Influence", "Liquor_Laws", "Drunkenness", "Disorderly_Conduct", "Vagrancy", "All_Other_Offense", "Suspicion", "Curfew_Violation"]
crime_data.columns = crime_data.columns[:1].tolist() + namesList
crime_data
```

```python
# Slice rows under 18 data
crime18_df = crime_data[crime_data["Type"] == "Under 18"].reset_index(drop=True)
crime18_df.drop(crime18_df.columns[[1]], inplace=True, axis=1) # Drop type
crime18_df.count()
```

```python
filepath = os.path.join(".", "Resources/State/Crimes")

# Read all files from directory filepath
xList = [x for x in os.listdir(filepath)]
```

```python
# Loop thru the file list and write info to opened text file if criteria are met
i = 1

for x in xList: 
    # If error, record exception
    # Split the extension from the path and normalise it to lowercase.
    ext = os.path.splitext(x)[-1].lower()
    excel = os.path.splitext(x)[0].lower()
    
    if "Table_69_Arrest" in x:        
        if ext == ".xls" or ext == ".xlsx": 
            yr = excel.split("_")[5]
            print(yr)
            #Clean data by skipping and dropping all not needed columns and rows 
            crime_data = pd.read_excel(f'{filepath}/{x}', skiprows=3)
            crime_data["State"] = crime_data["State"].str.replace('\d+', '')  # Remove all digit
            crime_data.drop(crime_data.columns[[33,34]], inplace=True, axis=1) # Drop last 2 columns
            
            # Some sheets missing a state 
            if crime_data["State"].count() == 60:
                crime_data = crime_data.drop(crime_data.index[102:]) # Drop last 7 rows 
            else:
                crime_data = crime_data.drop(crime_data.index[100:]) # Drop last 7 rows 
            
            #Copy odd # row state col to even # row null state col
            for index, row in crime_data.iterrows():
                if index == 0:
                    state = row["State"]
                else:
                    if index % 2 == 0:
                        state = row["State"]
                    else:
                        crime_data["State"][index] = state            

            crime_data["State"] = crime_data.State.str.capitalize()   # Format state to readable
            # rename all cols
            namesList = ["Type", yr + "_Total_Crimes", "Violent", "Property", "Murder", "Rape", "Robery", "Assault", "Burglary", "Theft", "Vehicle_Theft", "Arson", "Other_Assault", "Forgery", "Fraud", "Embezzlement", "Stolen_Property", "Vandalism", "Weapons", "Prostitution", "Sex_Offense", "Drug_Abuse", "Gambling", "Family_Offenses", "Driving_Influence", "Liquor_Laws", "Drunkenness", "Disorderly_Conduct", "Vagrancy", "All_Other_Offense", "Suspicion", "Curfew_Violation"]
            crime_data.columns = crime_data.columns[:1].tolist() + namesList
            
            # Select only needed cols
            crime_data = crime_data[["State", "Type", yr + "_Total_Crimes"]]
            
            
            # Slice rows total age data
            crime_df = crime_data[crime_data["Type"] != "Under 18"].reset_index(drop=True)
            crime_df.drop(crime_df.columns[[1]], inplace=True, axis=1) # Drop type         

            # -------------------------------------------------------------------------------------------
            # Merge data by state
            print(f'{yr}, {crime_df["State"].count()}')
            if i == 1:
                crime_ttl_df =  pd.DataFrame(crime_df)
            else:
                if crime_df["State"].count() == 51 and crime_ttl_df["State"].count() == 50:
                    crime_ttl_df = pd.merge(crime_ttl_df, crime_df, how='right', on=['State'])
                elif crime_df["State"].count() == 50 and crime_ttl_df["State"].count() == 50:
                    crime_ttl_df = pd.merge(crime_ttl_df, crime_df, how='outer', on=['State'])
                elif crime_df["State"].count() == 50 and crime_ttl_df["State"].count() == 51:
                    crime_ttl_df = pd.merge(crime_ttl_df, crime_df, how='left', on=['State'])
                else:
                    crime_ttl_df = pd.merge(crime_ttl_df, crime_df, how='inner', on=['State'])
                
            i += 1    # Increment by 1 as excel sheet is cleaned and formatted

crime_ttl_df
```
